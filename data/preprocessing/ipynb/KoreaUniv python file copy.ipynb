```python
path = 'D:\AIHub\이상행동 CCTV 영상/01.폭행(assult)'
save_path = 'D:\AIHub\이상행동 CCTV 영상/downscaled/'
```

```python
import os
import glob
import shutil
```

```python
os.listdir(path)
```

```python
count = 0
for file in glob.glob(path +'/**/*_resized_720.mp4', recursive=True):
    #print(file, os.path.isdir(file))
    #if os.path.isdir(file):
    #    os.mkdir(save_path+file.split(path)[-1])
    print(file)
    out = save_path + file.split(path)[-1]
    #print(out)
    if not os.path.isfile(out):
        shutil.copy(file, out)
        count += 1
        
print('{} files added'.format(count))
```
