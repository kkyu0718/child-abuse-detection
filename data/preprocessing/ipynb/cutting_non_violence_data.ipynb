```python
import glob
import numpy as np
import os
import cv2
import shutil
```

```python
from moviepy.tools import subprocess_call
from moviepy.config import get_setting

def ffmpeg_extract_subclip(filename, t1, t2, targetname=None):
    """ Makes a new video file playing video file ``filename`` between
    the times ``t1`` and ``t2``. """
    name, ext = os.path.splitext(filename)
    if not targetname:
        T1, T2 = [int(1000*t) for t in [t1, t2]]
        targetname = "%sSUB%d_%d.%s" % (name, T1, T2, ext)

    cmd = [get_setting("FFMPEG_BINARY"),"-y",
           "-ss", "%0.2f"%t1,
           "-i", filename,
           "-t", "%0.2f"%(t2-t1),
           "-vcodec", "copy", "-acodec", "copy", targetname]

    subprocess_call(cmd)
```

```python
mp4_list = glob.glob(aihub_nofi_path+'/**/*.mp4', recursive=True)
```

```python
def get_partition(video_path, vid_limit = 17):
    cap = cv2.VideoCapture(video_path)
    fps = cap.get(cv2.CAP_PROP_FPS)      
    frame_count = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    duration_limit = frame_count/fps

    partition = list()
    vid_len_cnt = 0.0

    while len(partition) < vid_limit and vid_len_cnt < duration_limit:
        length = round(np.random.rand() * 2 + 3, 2)
        partition.append(vid_len_cnt)
        vid_len_cnt += length
    return partition
```

```python
mp4_list = sorted(mp4_list)
```

```python
count = 0
for video in mp4_list:
    partition = get_partition(video)
    basename = os.path.basename(video).split('.mp4')[0]

    prev = partition[0]
    index = 0
    for next in partition[1:]:       
        index += 1
        save_name = save_path + '/' + basename + '_' + str(index) + '.mp4'
        ffmpeg_extract_subclip(video, prev, next, targetname=save_name)
        #print(index, prev, next) # cut here
        prev = next
    count += 1
    print('{}/{} done ------- {} %'.format(count, len(mp4_list), round(count/len(mp4_list) * 100, 3)))
print('done !')
```

```python
yt_vid_list = glob.glob(yt_nofi_path + '/**/*.mp4', recursive=True)
```

```python
def youtube_get_partition(video_path, vid_limit = 17):
    cap = cv2.VideoCapture(video_path)
    fps = cap.get(cv2.CAP_PROP_FPS)      
    frame_count = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    duration_limit = frame_count/fps
    
    partition = list()
    vid_len_cnt = 0.0
    
    while len(partition) < vid_limit and vid_len_cnt < duration_limit:
        length = round(np.random.rand() * 2 + 3, 2)
        partition.append(vid_len_cnt)
        vid_len_cnt += length
    return partition
```
